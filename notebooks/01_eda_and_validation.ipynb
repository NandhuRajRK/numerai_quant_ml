# Numerai EDA and Validation

This notebook is intentionally lightweight. Use it after running the scripts to inspect feature distributions, era sizes, validation correlations, and prediction files without turning notebook state into the source of truth.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
DATA_DIR = PROJECT_ROOT / "data" / "raw" / "v5.2"
PRED_DIR = PROJECT_ROOT / "data" / "predictions"
```

## Load a Small Sample

The full training file can be large. Start with a few columns while exploring.

```python
train_path = DATA_DIR / "train.parquet"
if train_path.exists():
    preview = pd.read_parquet(train_path).head(10_000)
    display(preview.head())
    print(preview.shape)
else:
    print("Run: uv run python scripts/download_data.py")
```

```python
if "preview" in globals():
    era_counts = preview["era"].value_counts().sort_index()
    era_counts.plot(kind="bar", figsize=(12, 3), title="Rows per era in preview")
    plt.tight_layout()
```

## Inspect Validation Diagnostics

After running `scripts/validate_baseline.py`, inspect the era correlation distribution.

```python
era_corr_path = PRED_DIR / "validation_era_correlations.csv"
if era_corr_path.exists():
    era_corr = pd.read_csv(era_corr_path)
    display(era_corr.describe())
    era_corr["corr"].plot(kind="hist", bins=30, title="Validation era correlations")
else:
    print("Run: uv run python scripts/validate_baseline.py")
```
